```python
#filter for the races based in USA

df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combining all filters
```

```python
AR = df[(df['Event distance/length'].isin(['50km','50mi'])) &  (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')] 
```

```python
AR.head(10)
```

```python
#AR means Athletic race
AR.shape
```

```python
#removing the (USA) from event name
```

```python
AR['Event name']. str.split('(').str.get(0)
```

```python
AR['Event name'] = AR['Event name']. str.split('(').str.get(0)
```

```python
AR.head(10)
```

```python
#clean up athlete age
#To get the age, we are going to do 'the year of the event' - 'the athletes year of birth'. Basically creating a new column
```

```python
AR['athlete_age'] = 2020 - AR['Athlete year of birth']
```

```python
AR['athlete_age']
```

```python
#REMOVING h from athlete performance
AR['Athlete performance'] = AR['Athlete performance']. str.split('').str.get(0)
```

```python
AR['Athlete performance']
```

```python
AR.head(5)
```

```python
#drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
```

```python
#axis = 1 shows that we are only dropping the column
AR = AR.drop(['Athlete club', 'Athlete country' , 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
AR.head()
```

```python
#clean up any null values
AR.isna().sum()
#tells us where the null values are
```

```python
AR[AR['athlete_age'].isna()==1]
#basically, where all the values ==1 return NA for True
```

```python
AR = AR.dropna()
#drop the above rows
```

```python
AR.shape
```

```python
#checking for duplications
AR[AR.duplicated() == True]
```

```python
#reset index
AR.reset_index(drop = True)
```

```python
#fix types
AR.dtypes
```

```python
AR['athlete_age'] = AR['athlete_age'].astype(int)
#changing the type to int
```

```python
AR['athlete_age'].dtype
```

```python
AR['Athlete average speed'] = AR['Athlete average speed'].astype(float)
```

```python
AR['Athlete average speed'].dtype
```

```python
AR.head()
```

```python
#renaming column names
AR = AR.rename(columns = {'Year of event':'year' , 'Event dates': 'race_date', 'Event name': 'race_name', 'Event distance/length': 'race_distance', 
                          'Event number of finishers': 'finishers', 'Athlete performance': 'performance', 'Athlete gender': 'gender',
                          'Athlete average speed': 'average_speed', 'Athlete ID': 'ID', 'athlete_age': 'age'
    
    
    
    
    
    
    
    
    
    
    
})
```

```python
AR.head()
```

```python
#reordering columns
AR2 = AR[['race_date','race_name', 'race_distance','finishers', 'ID', 'gender', 'age', 'year', 'average_speed']]
```

```python
#what age groups are the best in the 50m race (20 + races min)
```

```python
AR2.query('race_distance == "50mi"').groupby('age')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9').head(20)
```

```python
#seasons for the data -> Slower in summer than winter?
#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split to 2 decimals
```

```python
AR2['race_month'] = AR2['race_date'].str.split('.').str.get(1).astype(int)
```

```python
#creating the column for seasons
AR2['race_season'] = AR2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
AR2.head(25)
```

```python
AR2.groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
#50miler only
AR2.query('race_distance =="50mi"').groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
